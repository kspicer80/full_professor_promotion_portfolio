```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import seaborn as sns
```

```python
file_path = Path(r"G:\My Drive\Shared English Department Faculty Folder\assessment_data_&_reports\assessment_databank\main_assessment_data_file_updated_may_27_2023.xlsx")
df = pd.read_excel(file_path)
df = df.drop(columns=['student_name'])
df.head()
```

```python
course_filter = df[df['instructor'] == 'Spicer']
course_filter.head()
```

```python
course_filter['year'] = course_filter['year'].astype(int)
course_filter['score'] = course_filter['score'].astype(float)
```

```python
course_filter.dtypes
```

```python
course_filter['score'] = pd.to_numeric(course_filter['score'], errors='coerce')
rows_with_nan =  course_filter[course_filter['score'].isna()]
rows_with_nan
```

```python
course_filter['year'] = course_filter['year'].astype(str)
course_filter['semester_year'] = course_filter['semester'] + ' ' + course_filter['year']
```

```python
rows_with_whitespace = course_filter[course_filter['semester_year'].str.contains('\s')]
rows_with_whitespace
```

```python
# Assuming your DataFrame is named 'df' with columns 'semester_year' and 'outcomes' and 'scores'
grouped = course_filter.groupby('semester_year')['score'].mean()
print(grouped)
```

```python
desired_order = ['Fall 2018', 'Spring 2019', 'Fall 2019', 'Spring 2020', 'Summer 2020', 'Fall 2020', 'Spring 2021', 'Fall 2021', 'Spring 2022', 'Fall 2022', 'Spring 2023']
grouped_sorted = grouped.loc[desired_order]
```

```python
ax = grouped_sorted.plot(kind='bar', color='green')
plt.xlabel('Semester/Year')
plt.ylabel('Mean Score')
plt.title('Mean Scores by Semester/Year for Upper-Division English Courses')

#for i, v in enumerate(grouped_sorted):
    #ax.text(i, v, f'{v:.2f}', ha='center', va='bottom')
plt.show()
```

```python
# Group by 'semester_year' and 'outcome' to calculate the mean scores
grouped = course_filter.groupby(['semester_year', 'outcome'])['score'].mean().reset_index()

#custom_palette = sns.color_palette("Paired", 18)

# Plotting the scatter plot
sns.scatterplot(data=grouped, x='semester_year', y='score', hue='outcome', alpha=1, palette='tab20')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Semester/Year')
plt.ylabel('Mean Score')
plt.title('Mean Scores on English Dept. Outcomes by Semester/Year')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
