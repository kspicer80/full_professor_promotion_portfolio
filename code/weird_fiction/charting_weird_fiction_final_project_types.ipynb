```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"G:\My Drive\DHStuff\projects\weird_fiction_final_projects_playground\student_project_types.json"
```

```python
df = pd.read_json(file_path)
counts = df['category'].value_counts()
```

```python
counts.plot(kind='bar')
plt.xlabel('project type')
plt.ylabel('count')
plt.title("counts by project type")
plt.show()
```

```python
sns.barplot(x=counts.index, y=counts.values)
plt.xlabel('project type')
plt.ylabel('count')
plt.title("counts by project type")
plt.xticks(rotation=90)
plt.show()
df.head()
```

```python
counts = df.groupby(['category', 'subfolder']).size().reset_index(name='count')

# Plot the counts as a bar chart using seaborn
plt.figure(figsize=(20, 10))
#sns.color_palette('plasma', as_cmap=True)
sns.color_palette('tab10')
sns.barplot(x='category', y='count', hue='subfolder', data=counts, palette='tab10', width=5, dodge=50)
plt.xlabel('project type')
plt.ylabel('counts')
plt.title('counts by project type and semester/year')

# Adjust the legend position
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=90)
plt.xticks(fontsize=12)
plt.show()
```

```python
g = sns.stripplot(x='category', y='count', hue='subfolder', data=counts, jitter=25, dodge=True, palette='inferno')

# Customize the plot
g.figure.set_size_inches(12, 6)
plt.xlabel('Project Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.xticks(fontsize=12)
plt.title('Counts by Project Type and Semester/Year')
plt.legend(title="semester/year")
plt.show()
```
